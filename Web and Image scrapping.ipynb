### Image Scrapping for Wuzzuf Page 1

```python
from PIL import Image
from IPython.display import display

url = "https://wuzzuf.net/search/jobs/?q=&a=hpb" 
response = requests.get(url)
html_website = bs(response.text, "html.parser")

image_tags = html_website.find_all("img", class_="css-17095x3") 

for idx, image_tag in enumerate(image_tags, start=1):
    image_url = image_tag["src"]

    img_data = requests.get(image_url).content
    image_path = f"company_logo_{idx}.png"
    with open(image_path, "wb") as file:
        file.write(img_data)
        print(f"✅ Image {idx} saved as {image_path}")

    image = Image.open(image_path)
    display(image)


```
